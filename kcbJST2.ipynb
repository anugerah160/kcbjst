```python
import random

def activation_function(net, threshold):
    return 1 if net >= threshold else 0

def main():
    # Input miu dan threshold
    miu = float(input("Masukkan miu (Tingkat pembelajaran): "))
    threshold = float(input("Masukkan threshold: "))

    # Inisialisasi bobot dan bias secara acak
    w = [random.random(), random.random()]
    b = random.random() #w awal kali bias

    # Input dan output logika AND secara default
    input_data = [(0, 0), (0, 1), (1, 0), (1, 1)]
    output_data = [0, 0, 0, 1]

    print("Data Logika AND:")
    for i in range(len(input_data)):
        print("x1={}, x2={} output={}".format(input_data[i][0], input_data[i][1], output_data[i]))

    # Pelatihan model
    for epoch in range(100):  # Jumlah iterasi

        total_error = 0

        for i in range(len(input_data)):
            # Perhitungan net input
            net = w[0] * input_data[i][0] + w[1] * input_data[i][1] + b

            # Perhitungan output dengan fungsi aktivasi
            output = activation_function(net, threshold)

            # Perhitungan error
            error = output_data[i] - output
            total_error += abs(error)

            # Update bobot dan bias
            w[0] += miu * error * input_data[i][0]
            w[1] += miu * error * input_data[i][1]
            b += miu * error

        # Cetak bobot, output, dan total error setiap epoch
        print("\nEpoch:", epoch + 1)
        print("Bobot (w1, w2):", w)
        print("Bias (b):", b)
        print("Detail Output:")
        for i in range(len(input_data)):
            net = w[0] * input_data[i][0] + w[1] * input_data[i][1] + b
            output = activation_function(net, threshold)
            print("x1={}, x2={} output={}".format(input_data[i][0], input_data[i][1], output))
        print("Total Error:", total_error)

        # Cek konvergensi
        if total_error == 0:
            break

if __name__ == "__main__":
    main()
```

```python
import random

def activation_function(net):
    return 1 if net >= 0 else 0

def main():
    # Input miu (learning rate)
    miu = float(input("Masukkan miu (Tingkat pembelajaran): "))

    # Initialize weights and bias randomly
    w = [random.random(), random.random()]
    b = random.random() # Initial bias

    # Input and output for AND logic gate
    input_data = [(0, 0), (0, 1), (1, 0), (1, 1)]
    output_data = [0, 0, 0, 1]

    print("Data Logika AND:")
    for i in range(len(input_data)):
        print("x1={}, x2={} output={}".format(input_data[i][0], input_data[i][1], output_data[i]))

    # Train the model
    for epoch in range(100):  # Number of iterations
        total_error = 0

        for i in range(len(input_data)):
            # Calculate net input
            net = w[0] * input_data[i][0] + w[1] * input_data[i][1] + b

            # Calculate output using activation function
            output = activation_function(net)

            # Calculate error
            error = output_data[i] - output
            total_error += abs(error)

            # Update weights and bias
            w[0] += miu * error * input_data[i][0]
            w[1] += miu * error * input_data[i][1]
            b += miu * error

        # Print weights, output, and total error for each epoch
        print("\nEpoch:", epoch + 1)
        print("Bobot (w1, w2):", w)
        print("Bias (b):", b)
        print("Detail Output:")
        for i in range(len(input_data)):
            net = w[0] * input_data[i][0] + w[1] * input_data[i][1] + b
            output = activation_function(net)
            print("x1={}, x2={} output={}".format(input_data[i][0], input_data[i][1], output))
        print("Total Error:", total_error)

        # Check for convergence
        if total_error == 0:
            break

if __name__ == "__main__":
    main()
```
